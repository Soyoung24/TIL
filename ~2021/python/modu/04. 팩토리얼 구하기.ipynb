# 문제4 팩토리얼 구하기

1부터 n까지 연속한 정수의 곱을 구하는 알고리즘을 만들어 보세요.

```python
# 내가 푼 방법이랑 책이랑 똑같다.

def factorial(n):
    fac=1
    for i in range(1,n+1):
        fac*=i
    return fac
```

```python
factorial(4)
```

## 재귀 호출: 다시 돌아가 부르기
어떤 함수 안에서 자기 자신을 부르는 것

```python
def hello():
    print("hello")
    hello()
hello()
```

이 과정 영원히 반복  
재귀 호출 프로그램이 정상적으로 작동하려면 '종료 조건'이 필요하다.  
그렇지 않으면 RecursionError가 발생한다.   

```python
# 재귀호출 알고리즘 이용  

def fact(n):
    if n<=1:
        return 1
    return n*fact(n-1)
```

```python
fact(1)
```

```python
fact(5)
```

1. n이 1 이하인지 비교한다.   
    1-1. 1이하는 아주 작아서 더는 계산하지 않아도 되는 '종료 조건' 이 때 결과값을 1로 돌려준다.    
2. n이 1보다 크면 n!=n(n-1)!이므로 n*fact(n-1)을 결과값으로 돌려준다.    
3. 여기서 fact(n-1)은 어떻게 실행될까?    
    3-1. 다시 종료조건에 해당하는지 확인    
    3-2. fact(n-2) 호출   
    ...    
    3-4. 결국 f(1)을 만나게 된다. 따라서 재귀호출이 영원히 반복되지 않고 결국 답을 얻는다.

# 연습문제 4-1
문제 1의 1부터 n까지의 합 구하기를 재귀 호출로 만들어 보세요

```python
def sum_n(n):
    if n<1:
        return 0
    return n+sum_n(n-1)
```

```python
sum_n(10)
```

# 연습문제 4-2 
문제 2의 숫자 n개 중에서 최댓값 찾기를 재귀 호출로 만들어 보세요.

```python
def find_max(lst,n):
    if n==1:
        return lst[0]
    max_n_1 = find_max(lst, n - 1)  # n-1개 중 최댓값을 구함  
    if max_n_1>lst[n-1]:
        return max_n_1
    else:
        return lst[n-1]
```

```python
v = [17, 92, 18, 33, 58, 7, 33, 42]
print(find_max(v, len(v)))  # 함수에 리스트의 자료 갯수를 인자로 추가하여 호출
```
