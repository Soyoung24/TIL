```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/spotify/spotify.csv')
df.head()
```

### Q1 데이터는 현재 년도별 100곡이 인기순으로 정렬되어 있다. 각 년도별 1~100위의 랭킹을 나타내는 rank컬럼을 만들고 매년도 1위의 bpm컬럼의 평균값을 구하여라



```python
df['rank']=df.index%100+1
```

```python
df = df.dropna()
df[df['rank']==1]['bpm'].mean()
```

### Q2 2015년도에 가장많은 top100곡을 올린 artist는 누구인가?



```python
df[df['top year']==2015].artist.value_counts().index[0]
```

### Q3 년도별 rank값이 1~10위 까지의 곡들 중 두번째로 많은 top genre는 무엇인가?



```python
df[(df['rank']>=1)&(df['rank']<=10)]['top genre'].value_counts().index[1]
```

### Q4 피처링의 경우 title에 표시된다. 피처링을 가장 많이 해준 가수는 누구인가?



```python
df.head()
```

```python
def feat(row):
    if 'feat. ' in row:
        return row.split('feat. ')[1].split(')')[0]
    else:
        return None
df['feat']=df['title'].apply(feat)
```

```python
df['feat'].dropna().value_counts().index[0]
```

### Q5 top year 년도를 기준으로 발매일(year released)과 top100에 진입한 일자 (top year)가 다른 곡의 숫자를 count 했을때 가장 많은 년도는?



```python
df['not_same'] = df['year released']!= df['top year']  # 같지 않으면 true
```

```python
int(df.groupby('top year').not_same.sum().sort_values(ascending=False).index[0])
```

### Q6 artist 컬럼의 값에 대소문자 상관없이 q 단어가 들어가는 아티스트는 몇명인가?



```python
len(df[df['artist'].str.lower().str.contains('q')].artist.unique())
```

### Q7 년도 상관없이 전체데이터에서 1\~50위와 51\~100위간의 dur 컬럼의 평균값의 차이는?



```python
df[df['rank'].isin(range(1,51))].dur.mean() - df[df['rank'].isin(range(51,101))].dur.mean()
```

### Q8 title을 띄어쓰기 단어로 구분 했을때 가장 많이 나온 단어는 무엇인가? (대소문자 구분 x)



```python
word = []
def count_word(row):
    word.extend(row.split('(feat')[0].split())
df['title'].str.lower().apply(count_word)
```

```python
word
```

```python
dd=dict()
for i in word:
    if i in dd:
        dd[i]+=1
        
    else:
        dd[i]=1
[k for k, v in dd.items() if max(dd.values())==v][0]
```

### Q9 년도별 nrgy값의 평균값을 구할때 최대 평균값과 최소 평균값의 차이를 구하여라



```python
df.groupby('top year').nrgy.mean().sort_values().values[-1]- df.groupby('top year').nrgy.mean().sort_values().values[0]
```
